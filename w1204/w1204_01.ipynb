```python
browser = webdriver.Chrome()
url = "http://www.naver.com"
# 1. naver 페이지 열기
browser.get(url)
# 새페이지 열기
browser.get("http://www.daum.net")
# 뒤로가기
browser.back()
# 앞으로 가기
browser.forward()
browser.back()
# 네이버 로그인 클릭
browser.find_element(By.CLASS_NAME,'MyView-module__link_login___HpHMW').click()

# 아이디 선택
elem = browser.find_element(By.ID,'id')
elem.click()
# 아이디 입력
elem.send_keys("onulee")

# 패스워드 선택,입력
elem = browser.find_element(By.ID,'pw')
elem.click()
elem.send_keys("")

# 로그인버튼 클릭
browser.find_element(By.ID,'log.login').click()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
from bs4 import BeautifulSoup
import time
import os
import random
```

```python

url = "https://nid.naver.com/nidlogin.login?mode=form&url=https://www.naver.com/"
browser = webdriver.Chrome()
# 1. naver 페이지 열기
browser.get(url)

# input데이터 넣기
input_js = '''
document.getElementById("id").value = "{id}";
document.getElementById("pw").value = "{pw}";
'''.format(id='onulee',pw="j201")

time.sleep(3)
# time.sleep(random.uniform(1,5))
# naver에서 elem 찾기해서 데이터 입력시 차단
# 자바스크립트를 사용해서 데이터 입력
browser.execute_script(input_js)
time.sleep(3)
browser.find_element(By.ID,"log.login").click()
```

```python
# 탭 이동
browser.switch_to.window(browser.window_handles[0])
```

```python
# 메일 버튼 클릭
elem = browser.find_element(By.CLASS_NAME,'MyView-module__item_link___Dzbpq')
time.sleep(1)
elem.click()
browser.find_element(By.CLASS_NAME,'MyView-module__link_service___Ok8hP').click()
# 메일쓰기 클릭
browser.find_element(By.CLASS_NAME,'MyView-module__link_tool___tAoH1').click()
```

```python
# 탭 이동
browser.switch_to.window(browser.window_handles[1])

# 이메일주소 입력
elem = browser.find_element(By.ID,'recipient_input_element')
elem.send_keys('onulee@naver.com')

# 제목입력
elem = browser.find_element(By.ID,'subject_title')
elem.send_keys('날씨를 알려드립니다.')

# 내용입력
iframe = browser.find_element(By.TAG_NAME, "iframe")
browser.switch_to.frame(iframe)

# 요소 찾기
elem = browser.find_element(By.CSS_SELECTOR, ".workseditor-content")

# 텍스트 삽입
browser.execute_script(
    "arguments[0].innerText = '오늘 날씨가 좋아요.';", elem
)

# 보내기 버튼 클릭
browser.find_element(By.CLASS_NAME,'button_write_task').click()

```

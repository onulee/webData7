```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import requests
from bs4 import BeautifulSoup
import time
import os
import random
import csv
```

```python
url = "https://www.yeogi.com/"
browser = webdriver.Chrome() # 창열기
browser.maximize_window() # 최대창 확대
# 1. naver 페이지 열기

browser.get(url)
```

```python
# 강릉 입력
elem = browser.find_element(By.CLASS_NAME,"css-72v42m")
elem.send_keys("강릉")
# 날짜열기,날짜선택1,날짜선택2
elem = browser.find_element(By.XPATH,'//*[@id="__next"]/main/section[1]/div/div[1]/div[2]/div[2]/div[1]')
time.sleep(1)
elem.click()

elem = browser.find_element(By.XPATH,'//*[@id="__next"]/main/section[1]/div/div[1]/div[2]/div[2]/div[2]/div/div[1]/div[2]/div[1]/div/div/ul[2]/li[30]/button')
time.sleep(1)
elem.click()
elem = browser.find_element(By.XPATH,'//*[@id="__next"]/main/section[1]/div/div[1]/div[2]/div[2]/div[2]/div/div[1]/div[2]/div[1]/div/div/ul[2]/li[31]/button')
time.sleep(1)
elem.click()

# 검색버튼 클릭
elem = browser.find_element(By.XPATH,'//*[@id="__next"]/main/section[1]/div/div[1]/div[2]/div[4]/button')
time.sleep(1)
elem.click()
```

```python
# 화면스크롤 내림
pre_height = browser.execute_script("return document.body.scrollHeight")
while True:
    # 스크롤 아래로 내림
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    # 추가데이터를 불러오는동안 대기
    time.sleep(3)
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == pre_height:
        break
    # 같지 않으면
    pre_height = curr_height
    print("높이 변화 : ",curr_height)

print("스크롤 완료 : ",curr_height)    
   
```

```python
# 데이터 출력
soup = BeautifulSoup(browser.page_source,"lxml")

div = soup.find("div",{"class":"css-1jiha5s"})
ds = div.find_all("div",{"class":"gc-thumbnail-type-seller-card-wrapper css-1u8qly9"})
```

```python
f = open("yeogi.csv","w",encoding="utf-8-sig",newline="")
writer = csv.writer(f)

writer.writerow(['순번','링크','제목','평점','평가수','가격'])

# 반복문
for i,d in enumerate(ds):
    print(f"{i+1}.")
    try:
        img = d.img["src"]
        title = d.find("h3",{"class":"gc-thumbnail-type-seller-card-title css-1gxx2ac"}).text.strip()
        rating = float(d.find("span",{"class":"css-9ml4lz"}).text.strip())
        ratingCount = int(d.find("span",{"class":"css-oj6onp"}).text.strip().split(" ")[0].strip()[:-1].replace(",",""))
        price = int(d.find("span",{"class":"css-5r5920"}).text.strip().replace(",",""))
        if rating>=4.5 and ratingCount>=3000 and price<=200000:
            print(img)
            print(title)
            print(rating)
            print(ratingCount)
            print(price)
            print("-"*50)
            writer.writerow([i+1,img,title,rating,ratingCount,price])
    except:
        print("[ 예외처리 됨 ]") 

f.close()
print("프로그램 완료")           
```
